# Project Title: Secret Message Decoder
You’re building a Secret Message Decoder that helps a team decode hidden messages sent in plain text. The messages have mixed characters, spaces, and random numbers inserted to confuse readers. Your program should clean, transform, and analyze the text — without using in, if, or for.

Tasks:

Ask the user to input a secret message.

-Use len() and string methods (like .isalpha() ,lower(),upper()).

Convert the entire message to uppercase.

-Replace all spaces with underscores.

-Show only every 3rd character of the message.

-Use slicing: message[::3]

-Reverse the message and display it.

-Display the cleaned and transformed result.

```python
secret_msg=input("Enter your secret message: ")
uppercase=secret_msg.upper()
lowercase=secret_msg.lower()
print("secret message converted to uppercase: ", uppercase,"\nsecret message converted to lowercase: ", lowercase)
print("Replaced all spaces with underscore: ",secret_msg.replace(" ","_"))
print("Every 3rd charcter of the secret msg: ",secret_msg[::3],)
print("Reversed the message:",secret_msg[::-1])
print("Length of the secret message: ",len(secret_msg))
print("Secret message is alphabet- ",secret_msg.isalpha())
```

Tasks:
Create a tuple with fixed product categories like Electronics, Stationery, and Clothing.

Create a list of products that can be updated dynamically.

Ask the user to enter a new product and add it to the list.

Ask the user to input the index of a product to remove it from the list.

Display the final list of products and the product categories.


```python
# step 1 Create a tuple with fixed product categories like Electronics, Stationery, and Clothing.
categories=("Electronics","Stationery","Clothing")

#step 2 Create a list of products
products=["TV","T-shirt","Shoes"]

#step 3 Ask the user to enter a new product and add it to the list.
new_product=input("Enter a new product Name to add in List:")
products.append(new_product)
print(f'Updated products list:{products}')

#step4 Ask the user to input the index of a product to remove it from the list.
remove_product=input("Enter a Product to Remove:")
products.remove(remove_product)
print(f'Removed and Updated products list:{products}')

```

## **Case Study:- Manage Online Course Data**
**Instructions:**

- Create three lists of course enrollments (Python, SQL, Power BI).

- Use append(), insert(), and extend() to update them.

- Remove at least one course using remove() or pop().

- Display the first two and last two enrollments using slicing.

- Find total, max, and min enrollments using built-in functions.

- Sort the lists in ascending and descending order.

```python
# Step 1: Create three lists of course enrollments
python_course = [25, 30, 18]
sql_course = [22, 27, 35]
powerbi_course = [15, 20, 28]

# Step 2: Update lists using append(), insert(), and extend()
python_course.append(40)
print(f"Append:{python_course}")
sql_course.insert(1, 32)
print(f"Insert:{sql_course}")
powerbi_course.extend([25, 30])
print(f"Extend:{powerbi_course}")

# Step 3: Remove one enrollment using remove() or pop()
python_course.remove(18)
print(f"Remove:{python_course}")
sql_course.pop()
print(f"Pop:{sql_course}")

# Step 4: Display first two and last two enrollments using slicing
print("Python - First 2:", python_course[:2])
print("Python - Last 2:", python_course[-2:])
print("SQL - First 2:", sql_course[:2])
print("SQL - Last 2:", sql_course[-2:])
print("Power BI - First 2:", powerbi_course[:2])
print("Power BI - Last 2:", powerbi_course[-2:])

# Step 5: Find total, max, and min enrollments using built-in functions
print(print("\nTotal Python enrollments:", sum(python_course)))
print("Max Python enrollment:", max(python_course))
print("Min Python enrollment:", min(python_course))

print("\nTotal SQL enrollments:", sum(sql_course))
print("Max SQL enrollment:", max(sql_course))
print("Min SQL enrollment:", min(sql_course))

print("\nTotal Power BI enrollments:", sum(powerbi_course))
print("Max Power BI enrollment:", max(powerbi_course))
print("Min Power BI enrollment:", min(powerbi_course))

# Step 6: Sort lists in ascending and descending order
python_course.sort()
print(f"\nSQL Ascending:{python_course}")

python_course.sort(reverse=True)
print(f"\nSQL Descending:{python_course}")

sql_course.sort()
print(f"\nSQL Ascending:{sql_course}")

sql_course.sort(reverse=True)
print(f"\nSQL Descending:{sql_course}")

powerbi_course.sort()
print(f"\nSQL Ascending:{powerbi_course}")

powerbi_course.sort(reverse=True)
print(f"\nSQL Descending:{powerbi_course}")
```
